```python
#install keras,tensorflow and imutils

import  numpy as np
import os
import matplotlib.pyplot as plt
from imutils import paths
```

```python
from tensorflow.keras.applications import MobileNetV2
```

```python
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import AveragePooling2D
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Input
from tensorflow.keras.models import Model
```

```python
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing.image import img_to_array
from tensorflow.keras.preprocessing.image import load_img
```

```python
from tensorflow.keras.utils import to_categorical
from sklearn.metrics import classification_report
```

```python
from sklearn.preprocessing import LabelBinarizer
from sklearn.model_selection import train_test_split
```

```python
Data=r'C:\Users\AAVEJ\Desktop\Mtech\MLP_project\DataSet3'
path=list(paths.list_images(Data))
```

```python
path
```

```python
data=[]
catogr=[]

for i in path:
    label=i.split(os.path.sep)[-2]
    catogr.append(label)
    image=load_img(i,target_size=(224,224))
    image=img_to_array(image)
    image=preprocess_input(image)
    data.append(image)
```

```python
data
```

```python
catogr #its our depedent variable
```

```python
data=np.array(data,dtype='float32') 
catogr=np.array(catogr)
```

```python
data.shape
```

```python
data
```

```python
catogr
```

```python
lb=LabelBinarizer()
catogr=lb.fit_transform(catogr)
catogr=to_categorical(catogr)
```

```python
catogr
```

```python
train_X,test_X,train_Y,test_Y=train_test_split(data,catogr,test_size=0.25,random_state=10,stratify=catogr)
```

```python
Img_generator=ImageDataGenerator(rotation_range=20,zoom_range=0.15,width_shift_range=0.2,height_shift_range=0.2,shear_range=0.15,horizontal_flip=True,vertical_flip=True,fill_mode='nearest')
```

```python
basemodel=MobileNetV2(weights='imagenet',include_top=False,input_tensor=Input(shape=(224,224,3)))
```

```python
basemodel.summary()
```

```python
headmodel=basemodel.output
headmodel=AveragePooling2D(pool_size=(7,7))(headmodel)
headmodel=Flatten(name='Flatten')(headmodel)
headmodel=Dense(128,activation='relu')(headmodel)
headmodel=Dropout(0.5)(headmodel)
headmodel=Dense(2,activation='softmax')(headmodel)
model=Model(inputs=basemodel.input,outputs=headmodel)
```

```python
for layer in basemodel.layers:
    layer.trainable=False
```

```python
model.summary()
```

```python
lr=0.0001
Epc=20
Bs=11

opt=Adam(lr=lr,decay=lr/Epc)
model.compile(loss='binary_crossentropy',optimizer=opt,metrics=['accuracy'])

U=model.fit(
      Img_generator.flow(train_X,train_Y,batch_size=Bs),
      steps_per_epoch=len(train_X)//Bs,
      validation_data=(test_X,test_Y),
      validation_steps=len(test_X)//Bs,
      epochs=Epc
)

model.save(r'C:\Users\AAVEJ\Desktop\Mtech\MLP_project\umar.model')
```

```python
predict=model.predict(test_X,batch_size=Bs)
predict=np.argmax(predict,axis=1)
print(classification_report(test_Y.argmax(axis=1),predict,target_names=lb.classes_))
```
